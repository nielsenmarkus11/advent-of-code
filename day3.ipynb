```python
import pandas as pd
```

```python
df = pd.read_csv('data/day3-input.txt', delimiter = r'(x|#|,|@|:)', header = None)
df = df[[2,4,6,8,10]]
df.columns = ['id','row','col','rowsize','colsize']
df['id'] = pd.to_numeric(df['id'])
df['row'] = pd.to_numeric(df['row'])
df['col'] = pd.to_numeric(df['col'])
df['rowsize'] = pd.to_numeric(df['rowsize'])
df['colsize'] = pd.to_numeric(df['colsize'])
```

```python
df.loc[1:2,:]
```

```python
all_squares = pd.DataFrame(columns=['id','row','col'])
for i,row in df.iterrows():
    for j in range(row['rowsize']):
        for k in range(row['colsize']):
            tmp_df = {'id': i, 'row': j+row['row'], 'col': k+row['col']}
            all_squares = all_squares.append(tmp_df, ignore_index=True)
```
